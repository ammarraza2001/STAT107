```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Fifa 23 Players Data.csv')
df
```

# Making your ideal FIFA23 Team with 150 Million Euros and Average Age <=25

### Selecting a GoalKeeper (Budget 10 Million)

```python
goalkeeper = df[(df['Potential'] >= 85) & (df['Best Position'] == "GK")]
goalkeeper_cheapest = goalkeeper.sort_values('Value(in Euro)', ascending=True)
goalkeeper_cheapest
```

```python
goalkeeper = goalkeeper_cheapest.iloc[[0]]
goalkeeper
```

### Selecting a Defence i.e. RB, LB, CB, and CB (Budget 60 Million)

```python
LB = df.loc[df['Positions Played'].str.contains("LB", case=False)]
LB_within_range = LB[(LB['Potential'] >= 85) & (LB['Overall'] >= 75) & (LB['Age'] <= 25)]
cheapest_LB = LB_within_range.sort_values('Value(in Euro)', ascending=True)
LB = cheapest_LB.iloc[[2]]
LB

```

```python
RB = df.loc[df['Positions Played'].str.contains("RB", case=False)]
RB_within_range = RB[(RB['Potential'] >= 85) & (RB['Overall'] >= 75) & (RB['Age'] <= 25)]
cheapest_RB = RB_within_range.sort_values('Value(in Euro)', ascending=True)
RB = cheapest_RB.iloc[[2]]
RB


```

```python
CB = df.loc[df['Positions Played'].str.contains("CB", case=False)]
CB_within_range = CB[(CB['Potential'] >= 85) & (CB['Overall'] >= 75) & (CB['Age'] <= 25)]
cheapest_CB = CB_within_range.sort_values('Value(in Euro)', ascending=True)
CB = cheapest_CB.iloc[0:2]
CB
```

### Selecting a MidField i.e. CDM, RM, LM, and CAM (Budget 60 Million) 

```python
CDM = df.loc[df['Positions Played'].str.contains("CDM", case=False)]
CDM_within_range = CDM[(CDM['Potential'] >= 85) & (CDM['Overall'] >= 75) & (CDM['Age'] <= 25)]
cheapest_CDM = CDM_within_range.sort_values('Value(in Euro)', ascending=True)
CDM = cheapest_CDM.iloc[[2]]
CDM
```

```python
LM = df.loc[df['Positions Played'].str.contains("LM", case=False)]
LM_within_range = LM[(LM['Potential'] >= 85) & (LM['Overall'] >= 75) & (LM['Age'] <= 25)]
cheapest_LM = LM_within_range.sort_values('Value(in Euro)', ascending=True)
LM = cheapest_LM.iloc[[4]]
LM
```

```python
RM = df.loc[df['Positions Played'].str.contains("RM", case=False)]
RM_within_range = RM[(RM['Potential'] >= 85) & (RM['Overall'] >= 75) & (RM['Age'] <= 25)]
cheapest_RM = RM_within_range.sort_values('Value(in Euro)', ascending=True)
RM = cheapest_RM.iloc[[2]]
RM
```

```python
CAM = df.loc[df['Positions Played'].str.contains("CAM", case=False)]
CAM_within_range = CAM[(CAM['Potential'] >= 85) & (CAM['Overall'] >= 75) & (CAM['Age'] <= 30)]
cheapest_CAM = CAM_within_range.sort_values('Value(in Euro)', ascending=True)
CAM = cheapest_CAM.iloc[[9]]
CAM
```

### Selecting Attack i.e. ST and ST (Budget Remaining: 31.6 Million)

```python
ST = df.loc[df['Positions Played'].str.contains("ST", case=False)]
ST_within_range = ST[(ST['Potential'] >= 85) & (ST['Overall'] >= 75) & (ST['Age'] <= 25)]
cheapest_ST = ST_within_range.sort_values('Value(in Euro)', ascending=True)
cheapest_ST_1 = cheapest_ST.iloc[[5]]

ST = df.loc[df['Positions Played'].str.contains("ST", case=False)]
ST_within_range = ST[(ST['Overall'] >= 80) & (ST['Age'] <= 32)]
cheapest_ST = ST_within_range.sort_values('Value(in Euro)', ascending=True)
cheapest_ST_2 = cheapest_ST.iloc[[0]]

ST = pd.concat([cheapest_ST_1, cheapest_ST_2])
ST
```

### Final Team

```python
df_team = pd.concat([goalkeeper, RB, LB, CB, CDM, LM, RM, CAM, ST], ignore_index=True)
df_team
```
